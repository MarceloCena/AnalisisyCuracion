Analizaremos un conjunto de tablas de crecimiento: altura y peso segun la edad para niños de 0 a 80 meses de edad.

```python
%matplotlib inline
import pandas as pd
import numpy as np
```

Leemos el __DATASET_1 = "lenageinf.csv"__

```python
Talla = pd.read_csv("../input/lenageinf.csv")
Talla.head(10)
```

```python
Talla.dtypes
```

***
Leemos el __DATASET_2 = "wtageinf.csv"__

```python
Peso = pd.read_csv("../input/wtageinf.csv")
Peso.head(10)
```

```python
Peso.dtypes
```

***
Graficamos las curvas correspondientes al peso para cada percentil para niños y niñas del  __DATASET_1__:

```python
indice = pd.Index(list(Talla.Sex))
indice = indice.get_loc("Sex")
droped = Talla[:indice]
sexo_percentil = droped.loc[:indice,['P3','P5','P10','P25','P50','P75','P90','P95','P97']]
sexo_percentil.astype(float).plot();
print('altura por edad (en meses) - Dataset=1, Sexo=Masculino')
```

```python
indice = pd.Index(list(Talla.Sex))
indice = indice.get_loc("Sex")
droped = Talla[indice+1:]
sexo_percentil = droped.loc[indice+1:,['P3','P5','P10','P25','P50','P75','P90','P95','P97']]
sexo_percentil.astype(float).plot();
print('Altura por edad (en meses) - Dataset=1, Sexo=Femenino')
```

***
Graficamos las curvas correspondientes al peso para cada percentil para niños y niñas del __DATASET_2__:

```python
filtrado_1 = Peso[Peso.Sex==1]
sexo_percentil = filtrado_1.loc[:,['P3','P5','P10','P25','P50','P75','P90','P95','P97']]
sexo_percentil.plot();
print('Pesos por edad (en meses) - Dataset=2, Sexo=Masculino')
```

```python
filtrado_2 = Peso[Peso.Sex==2]
sexo_percentil = filtrado_2.loc[:,['P3','P5','P10','P25','P50','P75','P90','P95','P97']]
sexo_percentil.plot();
print('Pesos por edad (en meses) - Dataset=2, Sexo=Femenino')
```
